```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
data.head()
```

```python
speed_mean = data["Sp. Atk"].mean()
def set_speed(val):
    if val < speed_mean:
        return "Low"
    elif val == speed_mean:
        return "Average"
    else:
        return "high"
```

```python
hp_mean = data["HP"].mean()
def set_hp(val):
    if val < hp_mean:
        return "Low HP"
    elif val == hp_mean:
        return "Average HP"
    else:
        return "High HP";
```

```python
data["Speed_high_low"] = data["Sp. Atk"].apply(set_speed)
```

```python
data["HP_high_low"] = data["HP"].apply(set_hp)
```
